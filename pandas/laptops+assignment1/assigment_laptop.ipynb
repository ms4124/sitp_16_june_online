```python
print("\nMissing values per column:")
print(df.isnull().sum())

if 'Rating' in df.columns:
    df['Rating'] = df['Rating'].fillna(df['Rating'].median())

df = df.dropna()
print("\nAfter handling missing values:", df.shape)
```

```python
for col in df.columns:
    if df[col].nunique() == 1:
        df.drop(col, axis=1, inplace=True)
        print(f"Dropped column with single value: {col}")

unwanted = ['Review Count', 'Rating Count']  # example: often similar / duplicate
df.drop(columns=[col for col in unwanted if col in df.columns], inplace=True)
```

```python
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("\nCategorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)
```
