```python
x_point=np.array([1,2,3,4,5])
y_point=np.array([6,7,8,9,1])
plt.plot(x_point,y_point) # plotting the points
plt.show()
```

```python
x_point=np.array([1,2,3,4,5])
y_point=np.array([6,4,9,7,1])
plt.plot(x_point,y_point,'o') # 'o' is the marker
plt.show()
```

```python
x_point=np.array([1,2,3,4,5])
y_point=np.array([6,7,8,9,1])
plt.plot(x_point,y_point,marker='o') #marker='o' adds a circle at each point
plt.show()
```

```python
x_point=np.array([1,2,3,4,5])
y_point=np.array([6,7,8,9,1])
plt.plot(x_point,y_point,marker='*') #marker='o' adds a circle at each point
plt.show()
```

```python
x_pt=np.array([1,4,6,7])
plt.plot(x_pt,'o:b') # plot the points
plt.show()
```

```python
x_pt=np.array([1,4,6,7])
plt.plot(x_pt,marker='o',ms=10) # marker='o' means circle
plt.show()
```

```python
x_pt=np.array([1,4,6,7])
plt.plot(x_pt,marker='*',ms=10,mec='y') # marker size 10, marker color yellow
plt.show()
```

```python
x_pt=np.array([1,4,6,7])
plt.plot(x_pt,marker='*',ms=10,mec='y',mfc='g') # marker size 10, yellow edge, green fill
plt.show()
```

```python
x_pt=np.array([1,4,6,7])
plt.plot(x_pt,ls='dotted')
plt.show()
```

```python
x_pt=np.array([1,4,6,7])
plt.plot(x_pt,ls='dashed')
plt.show()
```

```python
x_pt=np.array([1,4,6,7])
plt.plot(x_pt,ls='dashed',linewidth=5)
plt.show()
```

```python
x_point=np.array([1,2,3,4,5])
y_point=np.array([6,7,8,9,1])
plt.plot(x_point,) #marker='o' adds a circle at each point
plt.plot(y_point,)
plt.show()
```

```python
x_point=np.array([1,2,3,4,5])
y_point=np.array([6,7,8,9,1])
plt.plot(x_point,ls='dotted') # dotted line
plt.plot(y_point,ls='dashed') # dashed line
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
c = np.array(['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black', 'orange', 'red', 'green', 'blue', 'yellow', 'cyan'])
plt.scatter(x, y,alpha=0.5, s = sizes, c = c )
plt.show()
```
