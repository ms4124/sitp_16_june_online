### barplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([1,2,5,43,56,6,4,5,])
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([1,2,5,43,56,6,4,5,])
plt.bar(x,y,color="green")
plt.show()
```

### pie

```python
x=np.array([1,2,3,5,4])
plt.pie(x)
plt.show()
```

```python
x=np.array([1,2,3,5,4])
mylable=np.array(["aplle","banana","mango","orange","wateralon"])
myexplode=np.array([0,3,0,0,0])
plt.pie(x,labels=mylable,startangle=90,explode=myexplode)
plt.show()
```

```python
x=np.array([1,2,3,5,4])
mylable=np.array(["aplle","banana","mango","orange","wateralon"])
myexplode=np.array([0,3,0,0,0])
colour=np.array(["red","purple","pink","brown","green"])
plt.pie(x,labels=mylable,startangle=90,explode=myexplode,shadow=True,colors=colour)
plt.show()
```

```python
x=np.array([1,2,3,5,4])
mylable=np.array(["aplle","banana","mango","orange","wateralon"])
myexplode=np.array([0,3,0,0,0])
colour=np.array(["red","purple","pink","brown","green"])
plt.pie(x,labels=mylable,startangle=90,explode=myexplode,shadow=True,colors=colour,autopct='%1.1f%%')
plt.show()
```

```python
x=np.array([1,2,3,5,4])
mylable=np.array(["aplle","banana","mango","orange","wateralon"])
myexplode=np.array([0,3,0,0,0])
colour=np.array(["red","purple","pink","brown","green"])
plt.pie(x,labels=mylable,startangle=90,explode=myexplode,shadow=True,colors=colour,autopct='%1.1f%%')
plt.legend(title="fruits")
plt.show()
```
