```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data =sns.load_dataset("tips")
```

```python
data
```

```python
plt.plot(data['total_bill'])
plt.show()
```

```python
sns.lineplot(data['total_bill']) ##3 this is a line plot
```

```python
sns.scatterplot(x = 'total_bill', y ='tip', data = data )
```

```python
sns.histplot(data['total_bill']) ##3 this is a hist plot
```

```python
plt.hist(data['total_bill'])
```

##### kdeplot 

```python
sns.kdeplot(data['total_bill'])
```

#### catplot 
#### relplot

```python
sns.relplot(x = 'total_bill', y = 'tip', data = data)
```

```python
sns.relplot(x = 'total_bill', y = 'tip',hue = 'time', data = data)
```

```python
sns.relplot(x = 'total_bill', y = 'tip',hue = 'time',style = 'smoker', data = data)
```

```python
sns.relplot(x = 'total_bill', y = 'tip',hue = 'time',style = 'smoker',kind = 'scatter',col = "time", data = data)
```

```python
sns.relplot(x = 'total_bill', y = 'tip',hue = 'time',style = 'smoker',kind = 'scatter',col = "smoker", data = data)
```

```python
sns.relplot(x = 'total_bill', y = 'tip',hue = 'time',style = 'smoker',kind = 'scatter',col = "day", data = data)
```

```python
sns.relplot(y='total_bill',x='tip',data=data)
plt.title("this is relpot")
plt.xlabel("total_bill",fontsize=20,color="black")
plt.ylabel("tip",fontsize=20,color="pink")
plt.show()
```
